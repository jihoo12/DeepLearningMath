```python
import numpy as np

# Define a 3x3 matrix
A = np.array([[12, -51, 4], 
              [6, 167, -68], 
              [-4, 24, -41]])
# Perform QR Decomposition
Q, R = np.linalg.qr(A)

print("Q (Orthogonal Matrix):\n", Q)
print("\nR (Upper Triangular Matrix):\n", R)
```

```python
import numpy as np

# Define a 3x3 matrix
ssa = np.array([[1, 1, 1], 
              [1, 1, 1], 
              [1, 1, 1]])
# Perform QR Decomposition
Q, R = np.linalg.qr(ssa)

print("Q (Orthogonal Matrix):\n", Q)
print("\nR (Upper Triangular Matrix):\n", R)
```
