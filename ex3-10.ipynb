```python
from scipy.stats import norm
mean = 64.43
std_dev = 2.99
x = norm.cdf(69.348112344584905031297239474952220916748046875,mean,std_dev)
print(f"{x:.50f}")
```

```python
x=norm.ppf(0.95,mean,std_dev)
print(f"{x:.45f}")
```

```python
import random
for i in range(0,1000):
    random_p = random.uniform(0.0,1.0)
    random_weight = norm.ppf(random_p,loc=64.43,scale=2.99)
    print(random_weight)
```

```python
mean = 18
std_dev = 1.5
x = norm.cdf(21,mean,std_dev)-norm.cdf(15,mean,std_dev)
print(x)
```
