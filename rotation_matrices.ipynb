$$
R=
\begin{bmatrix}
\cos\theta & -\sin\theta
\\
\sin\theta & \cos\theta
\end{bmatrix}
$$

```python
import numpy as np
import matplotlib.pyplot as plt

def rotate_vector(vector, theta_deg):
    # 각도를 라디안으로 변환
    theta = np.radians(theta_deg)
    
    # 2x2 회전 행렬 R 정의
    R = np.array([
        [np.cos(theta), -np.sin(theta)],
        [np.sin(theta),  np.cos(theta)]
    ])
    
    # 행렬 곱 연산 (R @ v)
    return R @ vector

# 1. 초기 벡터 정의 (x축 방향 1,0)
v = np.array([1, 0])

# 2. 회전 각도 설정 및 변환 (45도 회전)
angle = 45
v_rotated = rotate_vector(v, angle)

# 3. 시각화
plt.figure(figsize=(6, 6))
plt.quiver([0, 0], [0, 0], [v[0], v_rotated[0]], [v[1], v_rotated[1]], 
           angles='xy', scale_units='xy', scale=1, color=['blue', 'red'], label=['Original', f'Rotated {angle}°'])

# 그래프 설정
plt.xlim(-1.5, 1.5)
plt.ylim(-1.5, 1.5)
plt.axhline(0, color='black', lw=1)
plt.axvline(0, color='black', lw=1)
plt.grid(True, linestyle='--')
plt.legend()
plt.title(f"Rotation Matrix Transformation (theta={angle}°)")
plt.show()
```

R_x(\theta)=

1 & 0 & 0
\\
0 & \cos\theta & -\sin\theta
\\
0 & \sin\theta & \cos\theta
\end{bmatrix}
$$

$$
R_y(\theta)= 
\begin{bmatrix}
\cos\theta & 0 & \sin\theta
\\
0 & 1 & 0
\\
-\sin\theta & 0 & \cos\theta
\end{bmatrix}

$$
R_z(\theta) = 
\begin{bmatrix}
\cos\theta & -\sin\theta & 0
\\
\sin\theta & \cos\theta & 0
\\
0 & 0 & 1
\end{bmatrix}
$$
