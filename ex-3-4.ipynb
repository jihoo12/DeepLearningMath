```python
sample = [0,1,5,7,9,10]
```

```python
def median(values):
    orderd = sorted(values)
    print(orderd)
    n = len(orderd)
    mid = int(n/2) - 1 if n %2==0 else int(n/2)
    if n %2 == 0:
        return (orderd[mid]+orderd[mid+1])/2.0
    else:
        return orderd[mid]
print(median(sample))
```
