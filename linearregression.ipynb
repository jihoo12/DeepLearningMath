$$y = wx + b$$
y(종속 변수):우리가 예측하고 싶은 결과물
<br>
x(독립변수):원인이 되는 데이터
<br>
w(가중치/기울기):x가 변화할때 y에 미치는 영향력
<br>
b(편향/절편):x가 0일때의 기본값

# 손실함수
MSE (평균 제곱 오차,Mean Squared Error)
<br>
Cross-Entropy (교차 엔트로피)

$y$: 실제 정답
<br>
$\hat{y}$:예측값

mse
$$MSE = \frac{1}{n} \sum_{i=1}^{n} (y_i - \hat{y}_i)^2$$

```python
#mse
import numpy as np

y_true = np.array([3, -0.5, 2, 7])
y_pred = np.array([2.5, 0.0, 2, 8])

# (실제값 - 예측값)의 제곱의 평균
mse = np.mean((y_true - y_pred) ** 2)
#[0.5,-0.5,0,-1]
#[0.25,0.25,0,1]
#(0.25+0.25+1+0)/4=0.375

print(f"MSE: {mse}")
```

BCE(Binary Cross Entropy)
$$BCE = -\frac{1}{n} \sum_{i=1}^{n} [y_i \log(\hat{y}_i) + (1 - y_i) \log(1 - \hat{y}_i)]$$

CCE(Categorical Cross Entropy)
$$CCE = -\frac{1}{n} \sum_{i=1}^{n} \sum_{j=1}^{m} y_{ij} \log(\hat{y}_{ij})$$

```python
#cross entropy
import numpy as np

def cross_entropy_loss(y_pred, y_true):
    """
    y_pred: 모델의 예측값 (확률 분포)
    y_true: 실제 정답 (One-hot 벡터)
    """
    # 로그 계산 시 0이 들어가서 -inf가 되는 것을 방지하기 위한 아주 작은 값(delta)
    delta = 1e-7
    return -np.sum(y_true * np.log(y_pred + delta))
# 예시 데이터 (3개의 클래스)
y_true = np.array([0, 1, 0])      # 정답은 2번째 클래스
y_pred = np.array([0.1, 0.7, 0.2]) # 모델의 예측 확률

loss = cross_entropy_loss(y_pred, y_true)
print(f"NumPy Loss: {loss:.4f}")
```
